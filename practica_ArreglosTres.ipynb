realizar con arreglos y con clase un menu donde capture la clave del producto, nombre del producto, costo del producto se le aplicara un 10%, para la existencia, venta y ganancia se iran cambiando dependiendo la existencia que haya, en dado caso de que ya no queden prodctos en existencia decir que ya no hay producto y cuando se pida un producto mostrara el producto con su clave

```python
class Producto:
    def __init__(self, clave, nombre, costo, existencia):
        self.clave = clave
        self.nombre = nombre
        self.costo = costo
        self.existencia = existencia
        self.descuento = 0.10
        self.ganancia_total = 0
        self.productos_vendidos = 0

    def aplicar_descuento(self):
        return self.costo - (self.costo * self.descuento)

    def vender(self, cantidad):
        if self.existencia >= cantidad:
            precio_venta = self.aplicar_descuento() * cantidad
            self.ganancia_total += precio_venta
            self.existencia -= cantidad
            self.productos_vendidos += cantidad
            print("Has vendido", cantidad, self.nombre + "(s).")
        else:
            print("No hay suficiente producto en existencia.")

    def mostrar_info(self):
        print("Clave:", self.clave, "| Nombre:", self.nombre, "| Costo (con 10% desc.):", round(self.aplicar_descuento(), 2),
              "| Existencia:", self.existencia, "| Ganancias:", round(self.ganancia_total, 2),
              "| Productos vendidos:", self.productos_vendidos)

def mostrar_menu():
    print("\n--- Menú ---")
    print("1. Agregar producto")
    print("2. Mostrar productos")
    print("3. Vender producto")
    print("4. Mostrar producto más caro y más barato")
    print("5. Salir")

def buscar_producto(clave, productos):
    for producto in productos:
        if producto.clave == clave:
            return producto
    return None

def mostrar_mas_caro_y_barato(productos):
    if productos:
        producto_mas_caro = productos[0]
        producto_mas_barato = productos[0]


        for producto in productos:
            if producto.costo > producto_mas_caro.costo:
                producto_mas_caro = producto
            if producto.costo < producto_mas_barato.costo:
                producto_mas_barato = producto

        print("\n--- Producto más caro ---")
        print("Nombre:", producto_mas_caro.nombre, "| Costo:", producto_mas_caro.costo)

        print("\n--- Producto más barato ---")
        print("Nombre:", producto_mas_barato.nombre, "| Costo:", producto_mas_barato.costo)
    else:
        print("No hay productos para mostrar.")


productos = []



while True:
    mostrar_menu()
    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        clave = input("Introduce la clave del producto: ")
        nombre = input("Introduce el nombre del producto: ")
        costo = float(input("Introduce el costo del producto: "))
        existencia = int(input("Introduce la cantidad en existencia: "))
        productos.append(Producto(clave, nombre, costo, existencia))
        print("Producto", nombre, "agregado exitosamente.")

    elif opcion == "2":
        print("\n--- Lista de productos ---")
        for producto in productos:
            producto.mostrar_info()

    elif opcion == "3":
        clave = input("Introduce la clave del producto a vender: ")
        producto = buscar_producto(clave, productos)
        if producto:
            cantidad = int(input("¿Cuántas unidades de " + producto.nombre + " deseas vender?: "))
            producto.vender(cantidad)
        else:
            print("Producto no encontrado.")

    elif opcion == "4":
        mostrar_mas_caro_y_barato(productos)

    elif opcion == "5":
        print("Saliendo del programa.")
        break

    else:
        print("Opción no válida. Intenta nuevamente.")
```
